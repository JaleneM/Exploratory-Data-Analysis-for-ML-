```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
ztest ,pval1 = stests.ztest(dataa['BG1M'], x2=dataa['GG1M'], value=0,alternative='two-sided')
print(float(pval1))
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
ztest ,pval1 = stests.ztest(dataa['BGUP_3'], x2=dataa['GGUP_3'], value=0.0,alternative='two-sided')
print(pval1)
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
